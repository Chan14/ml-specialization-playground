```python
import numpy as np
```

```python
def recat(pt, origins):
    """categorize a point based on distance from origin of clusters"""
    nclusters = len(origins)
    min_distance = 10000
    y_new = None
    for j in range(nclusters):
        temp = origins[j] - pt.reshape(
            2,
        )
        dist = np.sqrt(np.dot(temp.T, temp))
        if dist < min_distance:
            min_distance = dist
            y_new = j
    return y_new
```

```python
def recat_vectorized(pts, origins):
    diff = pts[:, np.newaxis, :] - origins
    dists = np.sqrt(np.sum(diff**2, axis=-1))
    return np.argmin(dists, axis=1)
```

```python
centers = np.array([[-1, 0], [1, 0], [0, 1], [0, -1], [-2, 1], [-2, -1]])
pt1 = np.array([-2, 1.2])
three_pts = np.array([[-2, 1.2], [-2, -0.8], [0, 0.5]])
```

```python
diffs = three_pts[:, np.newaxis, :] - centers
print(f"diffs = \n{diffs}")
dists = np.sqrt(np.sum(diffs**2, axis=-1))
print(f"dists = \n{dists}")
```

```python
r1 = recat(pt1, centers)
r2 = np.zeros(three_pts.shape[0])
for i in range(len(three_pts)):
    r2[i] = recat(three_pts[i], centers)
print(r1, r2)
r1 = recat_vectorized(pt1.reshape(1, -1), centers)
r2 = recat_vectorized(three_pts, centers)
print(r1, r2)
```

```python
def rbf_similarity(A, B, gamma):
    """
    A: (m, d) array
    B: (n, d) array
    gamma: scalar

    Returns:
        S: (m, n) matrix
        where S[i, j] = exp( -gamma * ||A[i] - B[j]||^2 )
    """
    m = A.shape[0]
    n = B.shape[0]
    S = np.zeros((m, n))

    for i in range(m):
        for j in range(n):
            diff = A[i] - B[j]
            dist_sq = np.dot(diff, diff)
            S[i, j] = np.exp(-gamma * dist_sq)

    return S
```

```python
A = np.array([[0.0, 0.0], [1.0, 0.0]])  # shape (2, 2)

B = np.array([[0.0, 1.0], [1.0, 1.0], [2.0, 0.0]])

gamma = 0.5

result = rbf_similarity(A, B, gamma)
print(result)
```

```python
diffs = A[:, np.newaxis, :] - B
dists1 = np.sum(diffs**2, axis=-1)
print(diffs)
print(dists1)
similarity_fn = lambda X: np.exp(-gamma * X)
print(similarity_fn(dists1))
```

```python
def rbf_similarity_vectorized(A, B, gamma):
    """
    A: (m, d) array
    B: (n, d) array
    gamma: scalar

    Returns:
        S: (m, n) matrix
        where S[i, j] = exp( -gamma * ||A[i] - B[j]||^2 )
    """
    diffs = A[:, np.newaxis, :] - B
    dists = np.sum(diffs**2, axis=-1)
    S = np.exp(-gamma * dists)
    return S
```
